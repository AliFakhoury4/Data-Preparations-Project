```python
rice = pd.read_csv('Rice_Cammeo_Osmancik.csv')

print(rice.isnull().sum())

rice
```

```python
def detect_outliers(data):
    outliers=[]

    threshold = 3
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)

    return outliers
```

```python
outliersArea = detect_outliers(rice["Area"])
outliersPerimeter = detect_outliers(rice["Perimeter"])
outliersMajorAxisLength = detect_outliers(rice["Major_Axis_Length"])
outliersMinorAxisLength = detect_outliers(rice["Minor_Axis_Length"])
outliersEccentricity = detect_outliers(rice["Eccentricity"])
outliersConvexArea = detect_outliers(rice["Convex_Area"])
outliersExtent = detect_outliers(rice["Extent"])

```
